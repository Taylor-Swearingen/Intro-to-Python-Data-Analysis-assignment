# Examine the data
First, we just wanted to examine our data and all of the variables we have available in our dataset. 

```python
# importing packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statistics as stats

# load in datasets
file1 = ("Apprentice_Chef_Dataset.xlsx")

# read the file
apprentice = pd.read_excel(io = file1)

# review apprentice
apprentice
```

# Remove non-numeric Data

```python
# refine the data to only numeric data to be used properly
# new dataset excluding names, emails, first name, & family name
selected_columns = apprentice[['REVENUE','CROSS_SELL_SUCCESS','TOTAL_MEALS_ORDERED', 'UNIQUE_MEALS_PURCH',
                                'CONTACTS_W_CUSTOMER_SERVICE', 'PRODUCT_CATEGORIES_VIEWED', 'AVG_TIME_PER_SITE_VISIT',
                                'CANCELLATIONS_BEFORE_NOON', 'CANCELLATIONS_AFTER_NOON', 'TASTES_AND_PREFERENCES',
                                'PC_LOGINS','MOBILE_LOGINS','WEEKLY_PLAN','EARLY_DELIVERIES', 'LATE_DELIVERIES',
                                'PACKAGE_LOCKER','REFRIGERATED_LOCKER','AVG_PREP_VID_TIME','LARGEST_ORDER_SIZE',
                                'MASTER_CLASSES_ATTENDED', 'MEDIAN_MEAL_RATING','AVG_CLICKS_PER_VISIT',
                                 'TOTAL_PHOTOS_VIEWED']]

# new dataframe with selected columns
apprentice_numeric = selected_columns.copy()

# check the new dataset
apprentice_numeric
```

In this cell we removed any variables that will not provide any value to our evaluation. The variables removed were: names (first and family name), emails, and phone numbers.

```python
# is there any data missing?
print(apprentice_numeric.isnull().any(axis = 0))
```

We check our numeric data frame to ensure there are no missing or blank variables or data points. Since the output shows False for all of the variables, we know that the data is complete and will not need to be modified to fill in any blanks.

# Evaluating the Distribution

```python
## creating a histogram plot for revenue ##

# setting figure size
fig, ax = plt.subplots(figsize = [10, 7])

#histogram function
sns.histplot(data  = apprentice_numeric,
             x     ='REVENUE',
             bins  = 'fd',
             kde   = True,
             color = 'black')

# add mean line
plt.axvline(x = apprentice_numeric['REVENUE'].mean(),
            color = 'red')

# add median line
plt.axvline(x = apprentice_numeric['REVENUE'].median(),
            color = 'blue')

# this adds a title
plt.title(label = "Distribution of Revenue")


# add an x-label
plt.xlabel(xlabel = 'Revenue')


# add a y-label
plt.ylabel(ylabel = 'Frequency')

# add a legend
plt.legend(labels =  ['distribution', 'mean', 'median'])

# formatting
plt.tight_layout()
plt.show()
```

This histogram shows the frequency of the revenue amounts from each customer. The distribution does not seem to be a normal or equally distributed shape. Any revenue above $6000 seems to be in the outer limits of the data given. Within the histogram we have a blue line representing the median, and a red line representing the mean (average) revenue seen. The median, being closer to the peak of the histogram distribution would be the best representation of the data, since the mean is going to be swayed by those outer limit data points. If there were any missing variables in our revenue variable from the above view, we would implement these into those missing areas to ensure our data would be complete and have the ability to analyze. Luckily we have a complete dataset for more accurate evaluation. 

# Examine the Correlations

```python
apprentice_correlation = apprentice_numeric.corr(method = 'pearson').round(decimals = 2)
apprentice_correlation
```

To see what variables had the most impact on each other we want to run a correlation matrix and review the results. Visually looking through the matrix above is a little difficult, so we produced a simplified version below. 

```python
## creating a colorful correlation matrix ##
# increase the size of the plot for better viewing
fig, ax = plt.subplots(figsize=(20,15))

sns.heatmap(data = apprentice_correlation, # name of the correlation matrix created above
            cmap       = 'coolwarm',   
            square     = True,        
            annot      = True, 
            linecolor  = 'black', 
            linewidths = 0.5)          

# creating a  title for the heatmap
plt.title("Correlation matrix for all variables in Apprentice Dataset\n")

#showing the plot
plt.show()
```

Our simplified matrix above is using only the numeric data frame that we created. We can see the variables most correlated (related) with revenue are: total meals ordered, average preparation video time, largest order size, master classes attended, median meal rating, average clicks per visit, and total photos viewed. The only variable out of the correlated that is negative is average clicks per visit. We know that the variable being negative means the higher the average clicks per visit the lower the revenue we get from that customer. The positively correlated variables mean the higher the variable to higher the revenue we have from that customer. The variables we will move forward with in our analysis are the average preparation video time, average clicks per visit, and  median meal rating.

# Revenue vs. Average Preparation Video Time

```python
# create a scatter plot for the most positive correlated variable
# prep vid time vs revenue
sns.lmplot(x          = 'AVG_PREP_VID_TIME',  
           y          = 'REVENUE',
           line_kws   = {'color':'green'},
           scatter    = True,       
           fit_reg    = True, 
           aspect     = 2,
           data       = apprentice_numeric)   


# formatting and displaying the plot
plt.title(label    = 'Revenue vs. Avg. Prep Video Time')
plt.xlabel(xlabel  = 'Avg. Prep Vid Time (minutes)')
plt.ylabel(ylabel  = 'Revenue')
plt.xlim(0, 600)
plt.tight_layout()
plt.show()
```

This is a scatter plot between the variables of average preparation video time and the revenue. As you can see from the green linear relationship line, these two variables have a positive relationship, meaning when one variable increases the other will increase as well. This is great to know when in relationship with revenue because we know that if we increase the preparation video time each customer averages, we will increase revenue. As with most datasets there are a few outliers within the scatter plot. The further we travel on each axis the more the data differs from the relationship line. 

# Revenue vs. Average Clicks per Visit

```python
# create a scatter plot for the most correlated variables
# clicks per visit vs revenue
sns.lmplot(x          = 'AVG_CLICKS_PER_VISIT',  
           y          = 'REVENUE',
           line_kws   = {'color':'red'},
           scatter    = True,       
           fit_reg    = True, 
           aspect     = 2,
           data       = apprentice_numeric)   


# formatting and displaying the plot
plt.title(label    = 'Revenue vs. Avg. Clicks per Visit')
plt.xlabel(xlabel  = 'Avg. Clicks Per Visit')
plt.ylabel(ylabel  = 'Revenue')
plt.xlim(0, 20)
plt.tight_layout()
plt.show()
```

The only variable that we see with a moderately negative relationship with revenue is the average clicks per visit. What we are observing is the higher the clicks per visit on the website the lower our revenue. This means the user experience on the website should be as simple as possible. To increase our revenue we need to lower the average number of clicks each customer has on their visit. My suggestion for this would be to create packages our pre-selected groupings of meals that customers can add to their carts with minimal effort. The customization options should still be available, but the easier the experience the more the customer will produce revenue for the company.

# Revenue vs Median Meal Rating

```python
# create a scatter plot for the most correlated variables
# median meal rating and revenue
sns.lmplot(x          = 'MEDIAN_MEAL_RATING',  
           y          = 'REVENUE',
           line_kws   = {'color':'green'},
           scatter    = True,       
           fit_reg    = True, 
           aspect     = 2,
           data       = apprentice_numeric)   


# formatting and displaying the plot
plt.title(label    = 'Revenue vs. Median Meal Rating')
plt.xlabel(xlabel  = 'Median Meal Rating')
plt.ylabel(ylabel  = 'Revenue')
plt.xlim(0.5, 5.5)
plt.tight_layout()
plt.show()
```

The best way to increase the median meal ratings, and the revenue, would be to take a deeper look into the meal ratings and see if there are specific and repetitive complaints about the meal packages. We can also look into any variables that are correlated with the median meal ratings, such as the average number of clicks per visit to the website.

# Median Meal Rating vs. Average Clicks per Visit

```python
#create a scatter plot for the most correlated variables
#median meal rating and avg clicks per visit
sns.lmplot(x          = 'MEDIAN_MEAL_RATING',  
           y          = 'AVG_CLICKS_PER_VISIT',
           line_kws   = {'color':'red'},
           scatter    = True,       
           fit_reg    = True, 
           aspect     = 2,
           data       = apprentice_numeric)   


# formatting and displaying the plot
plt.title(label    = 'Median Meal Rating vs. Avg. Clicks Per Visit')
plt.xlabel(xlabel  = 'Median Meal Rating')
plt.ylabel(ylabel  = 'Avg. Clicks Per Visit')
plt.xlim(0.5, 5.5)
plt.tight_layout()
plt.show()
```

Here we can see a strong negative relationship between average clicks per visit to the website and the median meal ratings. This is the strongest correlation we see between any of the variables in the filtered dataset. This relationship shows that the greater the average clicks per visit the lower the median meal ratings. This ties perfectly into our topic mentioned previously about decreasing the number of clicks our customers are averaging on the website. If we can create a more user friendly environment this would in turn raise the median meal ratings. This will also increase revenue!
